# Introducing Neural Networks

## What is the history of Neural Networks?

1940s - Neural networks were originally conceived. \
1960s - The concept of Backpropagation emerged. \
2010 - Neural Network began winning competitions.

## What is a Neuron?




A single artificial neuron is a computation unit, which performs:
- multiplication and summation operation
- and applies a function operator on the result

This is extreme simplification of biological neurons.

## What is a Neural Network?

A single neuron performs a simple calculation and by itself is not capable of learning complex functions.
But when combined with hundereds or thousands of other neurons, the interconnectvity results in relationships that approximate complex functions.

## What are parameters?
A parameter is a number which the neural network learns during the training.
Parameters are nothing but weights and biases of a NN.

## How do we calculate the number of parameters of a network?
Number of parameters is nothing but the sum of weights and biases of the network. 
1. Each neuron will have a bias term, except in the input layer. \
So the number of biases grow with the number of neurons in the hidden layers and the output layer.

2. Each connection between two neurons will have a weight term associated with it.



What are activation functions?


Give examples of activation functions.
